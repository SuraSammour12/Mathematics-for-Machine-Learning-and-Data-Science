# Systems of Linear Equations: Three Variables
## Goal

Understand how a system of three linear equations with three variables
can represent three planes in 3D space, and how their intersections define
a unique solution, no solution, or infinite solutions.

I will perform 3 experiments to visualize and analyze these systems.

| Experiment                                | Goal                                                    | Condition     |
| ----------------------------------------- | ------------------------------------------------------- | ------------- |
| 1ï¸âƒ£ Non-Singular System (Unique Solution) | Three planes intersect at one point                     | âœ… det(A) â‰  0  |
| 2ï¸âƒ£ Singular System (No Solution)         | Three planes never meet at one point (inconsistent)     | âš ï¸ det(A) = 0 |
| 3ï¸âƒ£ Singular System (Infinite Solutions)  | All planes intersect along a line or are the same plane | â™¾ï¸ det(A) = 0 |

## ðŸ§ªExperiment 1 â€” Non-Singular System (Unique Solution)
###  Mathematical Form


\begin{cases}
x + y + z = 6 \\
2y + 5z = -4 \\
2x + 5y - z = 27
\end{cases}


---

###  Matrix Form


A =
\begin{bmatrix}
1 & 1 & 1 \\
0 & 2 & 5 \\
2 & 5 & -1
\end{bmatrix}
x =
\begin{bmatrix}
x \\ y \\ z
\end{bmatrix}
b =
\begin{bmatrix}
6 \\ -4 \\ 27
\end{bmatrix}



```python
# ðŸ’» Code Implementation
import numpy as np

# Coefficient matrix A
A = np.array([
    [1, 1, 1],
    [0, 2, 5],
    [2, 5, -1]
])

# Constant vector b
b = np.array([6, -4, 27])

# Compute determinant
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

# Solve the system if non-singular
if det_A != 0:
    x = np.linalg.solve(A, b)
    print("Unique Solution [x, y, z]:", x)
else:
    print("The system is singular â€” no unique solution.")
```

### Output
Determinant of A: -21.0

Unique Solution [x, y, z]: [ 5.  3. -2.]


**Result:** The determinant is non-zero, so the system has a unique solution at
(x, y, z) = (5, 3, -2) â€” meaning the three planes intersect at a single point.


---
### ðŸ¤– Machine Learning Connection

Linear systems like this appear in Linear Regression, Least Squares, and 3D transformations.

When det(A) â‰  0, the model parameters are uniquely solvable.

When det(A) = 0, the data is linearly dependent (multicollinearity).

### ðŸŽ¨ 3D Plot: Visualize the three planes and their intersection

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # required for 3D plotting

# Grid for plotting
xx, yy = np.meshgrid(np.linspace(-2, 7, 30), np.linspace(-2, 7, 30))

# Plane equations rearranged to z = ...
# Eq1: x + y + z = 6       -> z = 6 - x - y
z1 = 6 - xx - yy

# Eq2: 0x + 2y + 5z = -4   -> z = (-4 - 2y) / 5
z2 = (-4 - 2*yy) / 5.0

# Eq3: 2x + 5y - z = 27    -> z = 2x + 5y - 27
z3 = 2*xx + 5*yy - 27

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# Plot surfaces with transparency
ax.plot_surface(xx, yy, z1, alpha=0.5, rstride=1, cstride=1, linewidth=0)
ax.plot_surface(xx, yy, z2, alpha=0.5, rstride=1, cstride=1, linewidth=0)
ax.plot_surface(xx, yy, z3, alpha=0.5, rstride=1, cstride=1, linewidth=0)

# Plot intersection point (5,3,-2)
ax.scatter(5, 3, -2, color='red', s=80, label='Intersection (5, 3, -2)')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Three Planes â€” Unique Intersection Point')
ax.legend()
plt.show()
```

### ðŸ§© Geometric Interpretation

In the 3D plot above:

- Each **colored plane** represents one of the three linear equations:
  1. \(x + y + z = 6\)
  2. \(2y + 5z = -4\)
  3. \(2x + 5y - z = 27\)
- The **red point (5, 3, -2)** is the **unique intersection point** of all three planes.  
  This point satisfies all equations simultaneously.
- Since all three planes meet at a **single point**, the system is **non-singular** â€”  
  meaning the coefficient matrix \(A\) is **invertible** ($\det(A) \ne 0$)
.
- Geometrically:
  - **Unique Solution â†’** the three planes intersect in exactly **one point**.
  - **No Solution â†’** planes are **parallel or inconsistent**.
  - **Infinite Solutions â†’** planes **coincide** (overlap).

##ðŸ§ª Experiment 2 â€” Singular System (No Solution)
### Goal:
To demonstrate a 3Ã—3 linear system where the planes are inconsistent â€”
they never intersect at a single point.

### Mathematical Form
$$
\begin{cases}
x + y + z = 6 \\
2x + 2y + 2z = 12 \\
x - y + z = 2
\end{cases}
$$

Note: In this case, plane 2 is a multiple of plane 1, so the system becomes inconsistent


---

### Matrix form
$$
A =
\begin{bmatrix}
1 & 1 & 1 \\
2 & 2 & 2 \\
1 & -1 & 1
\end{bmatrix},
\quad
x =
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix},
\quad
b =
\begin{bmatrix}
6 \\
12 \\
2
\end{bmatrix}
$$



```python
# ðŸ’» Code Implementation
import numpy as np

# Coefficient matrix A
A = np.array([
    [1, 1, 1],
    [2, 2, 2],
    [1, -1, 1]
])

# Constant vector b
b = np.array([6, 12, 2])

# Compute determinant
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

# Try to solve
try:
    x = np.linalg.solve(A, b)
    print("Solution [x, y, z]:", x)
except np.linalg.LinAlgError:
    print("The system is singular â€” no unique solution (inconsistent planes).")

```

### Output
Determinant of A: 0.0

The system is singular â€” no unique solution (inconsistent planes).




---
### ðŸŽ¨ Code: Plotting Three Inconsistent Planes


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid for x and y
x = np.linspace(-5, 10, 30)
y = np.linspace(-5, 10, 30)
X, Y = np.meshgrid(x, y)

# Define planes
Z1 = 6 - X - Y           # x + y + z = 6
Z2 = 12/2 - X - Y        # 2x + 2y + 2z = 12 â†’ same as x + y + z = 6
Z3 = 2 - X + Y           # x - y + z = 2

# Plot the planes
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z1, alpha=0.6, color='blue', label='Plane 1')
ax.plot_surface(X, Y, Z2, alpha=0.6, color='green', label='Plane 2')
ax.plot_surface(X, Y, Z3, alpha=0.6, color='orange', label='Plane 3')

# Labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Three Planes â€” No Common Intersection (Singular System)')

plt.show()
```

### ðŸ§© Geometric Interpretation

In the 3D plot above:

The blue and green planes are parallel (because they represent proportional equations).

The orange plane intersects them separately â€” but thereâ€™s no single point common to all three.

This means the system is inconsistent (no solution).

ðŸ”¹ Geometric meaning: the planes do not meet at a single point.

ðŸ”¹ Algebraic meaning: $\det(A) = 0$ â†’ matrix not invertible.



---


### ðŸ¤– Machine Learning Connection

In ML, a singular system means some features are linearly dependent.
For example, if one feature is a perfect multiple of another,
the matrix $$X^T X$$  in Linear Regression becomes non-invertible,
causing model training to fail or produce unstable coefficients.

## ðŸ§ª Experiment 3 â€” Singular System (Infinite Solutions)
### Mathematical Form
$$
\begin{cases}
x + 2y + 3z = 4 \\
2x + 4y + 6z = 8 \\
3x + 6y + 9z = 12
\end{cases}
$$

Notice that each equation is a scalar multiple of the first one, so all three planes are parallel or coincide â€” in this case, they coincide along a line (infinitely many intersection points).


---
### Matrix Form
Ax=b

where

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
2 & 4 & 6 \\
3 & 6 & 9
\end{bmatrix},
\quad
x =
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix},
\quad
b =
\begin{bmatrix}
4 \\
8 \\
12
\end{bmatrix}
$$


### Determinant Check

det(A)=0

This confirms that the matrix is singular, meaning there is no unique solution.

```python
# ðŸ’» Code Implementation
import numpy as np

# Coefficient matrix A (singular)
A = np.array([
    [1, 2, 3],
    [2, 4, 6],
    [3, 6, 9]
])

# Constant vector b
b = np.array([4, 8, 12])

# Compute determinant
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

# Check if singular or not
if det_A != 0:
    x = np.linalg.solve(A, b)
    print("Unique Solution [x, y, z]:", x)
else:
    print("The system is singular â€” infinitely many solutions or none.")
```

### ðŸŽ¨ 3D Plot (Geometric Interpretation)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting

# Create grid for plotting
xx, yy = np.meshgrid(np.linspace(-5, 5, 30), np.linspace(-5, 5, 30))

# Plane equations rearranged to z = ...
# Eq1: x + 2y + 3z = 4     -> z = (4 - x - 2y) / 3
z1 = (4 - xx - 2*yy) / 3

# Eq2: 2x + 4y + 6z = 8    -> z = (8 - 2x - 4y) / 6
z2 = (8 - 2*xx - 4*yy) / 6

# Eq3: 3x + 6y + 9z = 12   -> z = (12 - 3x - 6y) / 9
z3 = (12 - 3*xx - 6*yy) / 9

# Create 3D figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the three planes
ax.plot_surface(xx, yy, z1, alpha=0.6, color='red', rstride=1, cstride=1, linewidth=0)
ax.plot_surface(xx, yy, z2, alpha=0.6, color='blue', rstride=1, cstride=1, linewidth=0)
ax.plot_surface(xx, yy, z3, alpha=0.6, color='green', rstride=1, cstride=1, linewidth=0)

# Example point on the intersection line
# (any point that satisfies the equation, e.g., x=1, y=0, z=1)
ax.scatter(1, 0, 1, color='black', s=80, label='Point on Line of Intersection')

# Labels and legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Three Planes â€” Infinite Solutions (Singular System)')
ax.legend()

plt.show()
```

### ðŸ§© What the Plot Shows:

All three planes appear as one single flat surface (the dark overlapping plane you see).

Thatâ€™s because each equation is a multiple of the first â€” they represent the same plane in 3D space.

Mathematically:

Eq2 = 2Ã—Eq1, Eq3 = 3Ã—Eq1

â‡’ all describe the same plane.

There is no unique intersection point, because the three planes do not meet at one point.
Instead, every point on that plane satisfies all three equations.

The black point (1, 0, 1) plotted is simply an example of one of infinitely many points on the plane that satisfy the equations.

The label â€œPoint on Line of Intersectionâ€ is a bit symbolic here â€” in this specific case, the three planes completely coincide, so the â€œintersection lineâ€ degenerates into the entire plane itself.
