##üß™ Experiment 1 ‚Äî Non-Singular System (Unique Solution)

###  Theoretical Background  

Consider the linear system:  

$$
\begin{cases}
2x + y = 5 \\
x - y = 1
\end{cases}
$$  

This can be written in matrix form as:  

$$
A\mathbf{x} = \mathbf{b},
$$  

where  

$$
A =
\begin{bmatrix}
2 & 1 \\
1 & -1
\end{bmatrix},
\quad
\mathbf{b} =
\begin{bmatrix}
5 \\
1
\end{bmatrix}
$$  

The solution is given by:  

$$
\mathbf{x} = A^{-1} \mathbf{b},
\quad \text{if } \det(A) \neq 0
$$


````python
### üíª Code Implementation

import numpy as np

# Coefficient matrix A
A = np.array([[2,1],
              [1,-1]])

# Constant vector b
b = np.array([5,1])

# Compute the determinant
det_A = np.linalg.det(A)
print('Determinant of A:', det_A)

# Solve if determinant is nonzero
if det_A != 0:
  x = np.linalg.solve(A, b)
  print('Solution [x, y]:',x)
else:
  print('The system is singular - no unique solution.')
````

### Output
Determinant of A: -2.99

Solution [x, y]: [2. 1.]

### Result

The system has a unique solution:

x=2, y=1

### üé® Code: Plotting Two Linear Equations and Their Intersection

````python
import numpy as np
import matplotlib.pyplot as plt

# Equations:
# 1) 2x + y = 5   ‚Üí  y = 5 - 2x
# 2) x - y = 1    ‚Üí  y = x - 1

# Generate x values for plotting
x_vals = np.linspace(-2, 5, 100)

# Compute corresponding y values for each equation
y1 = 5 - 2*x_vals
y2 = x_vals - 1

# Intersection point (from previous solution)
x_sol, y_sol = 2.0, 1.0

# Create the plot
plt.figure(figsize=(6,6))

# Plot both lines
plt.plot(x_vals, y1, label='2x + y = 5', color='blue')
plt.plot(x_vals, y2, label='x - y = 1', color='green')

# Mark the intersection point
plt.scatter(x_sol, y_sol, color='red', s=80, label='Intersection (2, 1)')

# Add labels, title, legend, and grid
plt.xlabel('x')
plt.ylabel('y')
plt.title('Intersection of Two Linear Equations')
plt.legend()
plt.grid(True)

# Draw x and y axes for reference
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

# Display the plot
plt.show()
````

### Result:
The plot shows two straight lines intersecting at (2, 1) ‚Äî the unique solution of the system.



---

### ü§ñ Machine Learning Connection

In **Linear Regression**, the model parameters are computed as:

$$
\mathbf{w} = (X^{T}X)^{-1} X^{T} \mathbf{y}
$$

Here, $$X^{T}X$$ must be **invertible** (i.e., $$\det(X^{T}X) \neq 0$$) to obtain a **unique solution** for the model weights.  

If $$\det(X^{T}X) = 0$$, it means that the dataset‚Äôs features are **linearly dependent**, and the model cannot find **unique parameters**.

 Summary

- The lines intersect at (2, 1)

- The system is non-singular

##üß™ Experiment 2 ‚Äî Singular System (No Solution)

###  Goal
Explore a **singular system** where the two lines are **parallel** ‚Äî meaning they never intersect.  
In this case, the determinant of the coefficient matrix is **zero**, so the system has **no unique solution**.

---

###  Mathematical Form

Let‚Äôs define the system:

\
\begin{cases}
2x + y = 5 \\
4x + 2y = 8
\end{cases}


Matrix form:


A =
\begin{bmatrix}
2 & 1 \\
4 & 2
\end{bmatrix}
x =
\begin{bmatrix}
x \\
y
\end{bmatrix}
b =
\begin{bmatrix}
5 \\
8
\end{bmatrix}


---

###  Observation
Notice that the **second equation** is a multiple of the first in coefficients (each term √ó2),  
but **not** in constants (5 √ó 2 = 10 ‚â† 8).  
That means the two lines are **parallel**, not identical ‚Üí **no intersection point**.

---



````python
### üíª Code Implementation

import numpy as np
import matplotlib.pyplot as plt

# Coefficient matrix A
A = np.array([[2, 1],
              [4, 2]])

# Constant vector b
b = np.array([5, 8])

# Compute the determinant
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

# Check singularity
if det_A != 0:
    x = np.linalg.solve(A, b)
    print("Solution [x, y]:", x)
else:
    print("The system is singular ‚Äî no unique solution (parallel lines).")

````

###  Output

Determinant of A: 0.0
The system is singular ‚Äî no unique solution (parallel lines).




---

###  Result

The determinant of matrix **A** is zero (**det(A) = 0**), which means the system is **singular** and does **not have a unique solution**.  

Since both equations represent **parallel lines**, they never intersect ‚Äî therefore, **no solution exists**.

---

###  Geometric Interpretation

Each equation represents a straight line in the 2D plane:

(1) \( 2x + y = 5 \Rightarrow y = 5 - 2x \)  
(2) \( 4x + 2y = 8 \Rightarrow y = 4 - 2x \)

Both lines have the **same slope (-2)** but **different y-intercepts** (5 vs. 4).  
This means they are **parallel** and will **never intersect**.

Hence, the system is **singular** and has **no solution** ‚Äî the lines do not share any common point in the plane.




---

### üé® Code: Plotting Two Parallel Linear Equations (Singular System)


````python
import numpy as np
import matplotlib.pyplot as plt

# Equations:
# 1) 2x + y = 5  ‚Üí  y = 5 - 2x
# 2) 4x + 2y = 8 ‚Üí  y = 4 - 2x

# Generate x values
x_vals = np.linspace(-2, 5, 100)

# Compute y values for both lines
y1 = 5 - 2*x_vals
y2 = 4 - 2*x_vals

# Plot both lines
plt.figure(figsize=(6,6))
plt.plot(x_vals, y1, label='2x + y = 5', color='blue')
plt.plot(x_vals, y2, label='4x + 2y = 8', color='green')

# Formatting
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parallel Lines ‚Äî Singular System')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.show()
````

###  Result

The plot shows two **parallel straight lines** (blue and green) that **never intersect**.  
This means the system has **no unique solution**, since both equations represent lines with the **same slope** but **different intercepts**.

---

### ü§ñ Machine Learning Connection

In **Linear Regression**, model parameters are computed as:

$$
\mathbf{w} = (X^T X)^{-1} X^T \mathbf{y}
$$

For this inverse to exist,  

$$
\det(X^T X) \neq 0
$$  

must hold true.  

If  

$$
\det(X^T X) = 0,
$$  

it indicates that the features are **linearly dependent**, meaning some features are exact multiples of others.  
In such cases, the model **cannot compute unique weights** ‚Äî similar to how **parallel lines have no single intersection point**.

---

###  Summary

- Determinant = 0 ‚Üí lines are parallel ‚Üí **no unique solution**  
- The system is **singular**  
- Both equations represent the **same slope (-2)** but **different intercepts**  
- In machine learning, this corresponds to **redundant or dependent features** in the dataset

##üß™ Experiment 3 ‚Äî Singular System (Infinite Solutions)


###  Goal
Explore a **singular system** where the two lines are **identical** ‚Äî they perfectly overlap on the same line.  
In this case, the determinant is **zero**, but unlike the previous experiment, the system is **consistent** and has **infinitely many solutions**.

---
###  Mathematical Form


\begin{cases}
2x + y = 5 \\
4x + 2y = 10
\end{cases}


Matrix form:


A =
\begin{bmatrix}
2 & 1 \\
4 & 2
\end{bmatrix}
x =
\begin{bmatrix}
x \\
y
\end{bmatrix}
b =
\begin{bmatrix}
5 \\
10
\end{bmatrix}


---

###  Observation
Here, the **second equation** is exactly **2√ó the first**, including the constant term.  
That means both equations represent **the same line**.  
‚û°Ô∏è Every point that satisfies the first equation also satisfies the second.  
So the system has **infinitely many solutions** ‚Äî all the points on that line.

---

````python
### üíª Code Implementation

import numpy as np

# Coefficient matrix A
A = np.array([[2, 1],
              [4, 2]])

# Constant vector b
b = np.array([5, 10])

# Compute determinant
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

# Attempt to solve
if det_A != 0:
    x = np.linalg.solve(A, b)
    print("Unique Solution [x, y]:", x)
else:
    print("The system is singular ‚Äî infinitely many solutions (same line).")
````

### Output
Determinant of A: 0.0
The system is singular ‚Äî infinitely many solutions (same line).

---
### Result

Since det(A) = 0, the matrix is singular.
But because both equations describe the same line, the system is consistent and has infinite solutions.

In other words, there‚Äôs no single point ‚Äî every point on the line

y=5‚àí2x

is a valid solution.


---

### Geometric Interpretation

Both equations represent the same geometric line:

2x+y=5‚áíy=5‚àí2x

So instead of two different lines intersecting or running parallel,
we have one single line, counted twice.

---



### üé® Code: Plotting Two Identical Linear Equations (Singular System ‚Äî Infinite Solutions)


````python
import matplotlib.pyplot as plt

# Equations
# Both represent y = 5 - 2x
x_vals = np.linspace(-2, 5, 100)
y_vals = 5 - 2*x_vals

plt.figure(figsize=(6,6))
plt.plot(x_vals, y_vals, label='2x + y = 5 and 4x + 2y = 10', color='blue')

# Labels, title, legend, and grid
plt.xlabel('x')
plt.ylabel('y')
plt.title('Singular System ‚Äî Infinite Solutions (Same Line)')
plt.legend()
plt.grid(True)

# Draw axes
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)

plt.show()
````

### Result

The plot shows a single blue line,
confirming both equations are the same line, hence infinitely many (x, y) solutions exist.


---


### ü§ñ Machine Learning Connection

In Machine Learning, this situation happens when two features are perfectly correlated ‚Äî
one is an exact linear multiple of the other.

For example:

Feature2 = 2 √ó Feature1

This leads to multicollinearity, causing the feature matrix X·µÄX to become singular.
As a result, models like Linear Regression cannot compute


$$
(X^{T} X)^{-1}
$$


because the inverse doesn‚Äôt exist.

That‚Äôs why checking matrix rank or determinant before training a model is essential.



---


### Conclusion:
A singular but consistent system means both equations describe one line ‚Äî hence, infinite valid solutions.
