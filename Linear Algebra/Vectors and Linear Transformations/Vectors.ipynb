## The Norm of a Vector (‚Äñv‚Äñ)

The **norm** is the length (magnitude) of a vector.

It tells us *how long* the arrow is ‚Äî like the energy of a move.

image.png

Example:

image.png


```python
# üß™ Code Example
v = np.array([3, 4])
norm_v = np.linalg.norm(v)
print("‚Äñv‚Äñ =", norm_v)
```

### In ML:

The norm helps us measure distances in vector spaces, used everywhere from **regularization (L2 norm)** to **gradient magnitudes**.

If vectors are arrows, then adding them means following one arrow, then the other.

```python
# üß™ Code Example
u = np.array([1, 2])
v = np.array([3, 1])

sum_uv = u + v
diff_uv = u - v

print("u + v =", sum_uv)
print("u - v =", diff_uv)
```

## Distance Between Two Vectors

To measure how *different* two data points are, we use **Euclidean distance** ‚Äî the norm of their difference:

image.png



```python
# üß™ Code Example
u = np.array([1, 2])
v = np.array([3, 1])

distance = np.linalg.norm(u - v)
print("Distance between u and v =", distance)
```

### In ML:
Used in KNN, clustering, and any task that compares similarity or distance between examples.

## Multiplying a Vector by a Scalar

When you multiply a vector by a number (a **scalar**),

you scale its length ‚Äî stretch or shrink ‚Äî but keep its direction (unless the scalar is negative).

image.png

Example:

image.png

‚Üí same direction, twice as long.

```python
# üß™ Code Example
v = np.array([3, 4])
scaled_v = 2 * v
print("2 √ó v =", scaled_v)
```

### In ML:
This is like adjusting a gradient step:
a smaller learning rate = smaller vector (shorter step).

## Geometric Dot Product
The dot product can also be seen geometrically:

image.png

where ùúÉ is the angle between them.


- Œ∏=0¬∞ ‚Üí same direction ‚Üí large positive dot

- Œ∏=90¬∞ ‚Üí perpendicular ‚Üí dot = 0

- Œ∏=180¬∞ ‚Üí opposite ‚Üí large negative dot

```python
# üß™ Code Example
u = np.array([1, 0])
v = np.array([0, 1])

dot_uv = np.dot(u, v)
angle = np.arccos(dot_uv / (np.linalg.norm(u) * np.linalg.norm(v)))
print("u ¬∑ v =", dot_uv)
print("Angle (in degrees) =", np.degrees(angle))
```

### In ML:
Cosine similarity (used in NLP and recommender systems) is based on this relationship.

## Matrix √ó Vector Multiplication
Now that we understand vector scaling and addition, imagine applying many linear operations at once.

image.png

Each row of A transforms the input vector into a new value.

```python
# üß™ Code Example
A = np.array([[1, 2],
              [3, 4]])
x = np.array([5, 6])

Ax = np.dot(A, x)
print("A √ó x =", Ax)
```

### In ML:
Each neural network layer performs this operation ‚Äî multiplying inputs (vectors) by weights (matrices).
